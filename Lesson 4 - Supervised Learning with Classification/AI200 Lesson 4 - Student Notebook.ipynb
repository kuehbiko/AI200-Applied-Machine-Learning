<b>2b.iii) Train Model</b>

Now we can begin to create the Logistic regression. To do this, we make use of <b>scikit-learn (aka Sklearn)</b>. You can find Sklearn's documentations on linear regression in the link below:

- https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LogisticRegression.html#sklearn.linear_model.LogisticRegression

We start by training the model on the features. To do so:
- We first create a logistic regression object using the <code>LogisticRegression()</code> function
- On top of passing in the training dataset, We can pass in several default parameters:<br><br>
    - <code>max_iter</code> This sets the number of iterations which the model is allowed to try different parameters to find the maximum likelihood before it stops. If you do not specify any value, it takes on a values of default=100
Maximum number of iterations taken for the solvers to converge.<br><br>
- Thereafter we fit the model on the <b>training data</b> using the <code>fit()</code> function

```python
# train a logistic regression model on the training set
from sklearn.linear_model import LogisticRegression

# instantiate model
log_reg = LogisticRegression(max_iter=300)

# Earlier when trying to visualise the scatterplot we changed the outcome to str. Here we change it back!
insurance_diabetes_df['Outcome'] = insurance_diabetes_df['Outcome'].apply(int)

# Split data into predictors X and output Y
features = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin','BMI', 'DiabetesPedigreeFunction', 'Age']
X = insurance_diabetes_df[features] # Print x out if you don't understand what data it contains
y = insurance_diabetes_df['Outcome']

# fit model
log_reg.fit(X, y)
```

<b>2.iv) Generate predictions</b>

To generate predictions, we can use the <code>.predict()</code> function. In the example below, we are <b>simply making predictions on the dataset which the model was trained on.</b>

```python
log_reg.predict(X)
```

<div class="alert alert-block alert-warning">
Note that when you use the <code>predict()</code> function, the model automatically assumes a default threshold value of 0.5, and just classifies your observations based on that threshold. <br>

Also, it is important to note that in practice, you should <b>never train the model on the entire dataset (X and y), and subsequently use it to predict the values of the same dataset (X).</b> In the above example, we do so just to demonstrate the usage of <code>fit</code> and <code>predict</code> functions of the model.
    
Instead, you should remember to do a train-test split or K-Fold validation, and then <code>fit</code> the model on X_train and y_train, before evaluating the prediction output from using <code>predict</code> on X_test.
</div>

If you wish to set a threshold value, you would have to:
- Use the <code>predict_proba()</code> which will return you the estimated probability for each row of data that the outcome is positive in each instance  
- Write your own code to assign the outcome using your desired threshold for each row of data (see the two different ways below)

```python
y_pred_probabilities = log_reg.predict_proba(X)
y_pred_probabilities
```

<div class="alert alert-block alert-warning">
    
Let's seek to understand the above output:
    
- Each inner list corresponds to one row of observation in the dataset
- In each inner list the numbers sum to 1
- The two values in each list are:
    - <b>Value at positional index 0</b>: predicted probability that each observation is a member of class 0 (in the case of the example, it is the predicted probability that someone does not have diabetes)
    - <b>Value at positional index 1</b>:  predicted probability that each observation is a member of class 1  (in the case of the example, it is the predicted probability that someone has diabetes)
</div>

Given that the <b>Value at positional index 0</b> and <b>Value at positional index 1</b> essentially are referring to the same observation, we can just use one of them (usually we would use the value at positional index 1 as it is more intuitive)

```python
y_pred_probabilities_formatted = list(y_pred_probabilities[:, 1])
print(y_pred_probabilities_formatted)
```

Next, we just need to classify each predicted probability based on a threshold that is set by us. Below we show 2 methods for doing so:

<b>Method 1</b>

```python
y_pred = []

for pred in y_pred_probabilities_formatted:
    
    if pred > 0.6:
        
        y_pred.append(1)
        
    else:
        
        y_pred.append(0)
        
print(y_pred)
```

<b>Method 2</b>

```python
from sklearn.preprocessing import binarize

y_pred = binarize([y_pred_probabilities_formatted], threshold=0.6)[0]
print(y_pred)
```

<div class="alert alert-block alert-warning">
In the example above, we arbitrarily set the threshold to be 0.6. However, that is hardly the best way to determine our threshold value at all!
</div>

<b>2.v) Calculate Accuracy</b>

- Your predicted outcome

Let's see an example of how to use this function!

```python
from sklearn import metrics

# First we extract the outcomes from the original dataset
y_test = insurance_diabetes_df['Outcome']

# Then we pass in the test outcome and the predicted outcome to get the accuracy score
print(metrics.accuracy_score(y_test, y_pred))
```

<b>2b.vi) Confusion Matrix</b>

To create a confusion matrix, you just need to use the <code>metrics.confusion_matrix()</code> function. You will need to pass the function 2 parameters:
- The real outcome (from your test dataset)
- Your predicted outcome

Let's see an example of how to use this function!

```python
confusion = metrics.confusion_matrix(y_test, y_pred)
print(confusion)
```

Using the values from confusion matrix, we can easily extract the following:
- True Positives (TP): we correctly predicted that they do have diabetes
- True Negatives (TN): we correctly predicted that they don't have diabetes
- False Positives (FP): we incorrectly predicted that they do have diabetes (a "Type I error")
- False Negatives (FN): we incorrectly predicted that they don't have diabetes (a "Type II error")

```python
#[row, column]
TP = confusion[1, 1]
TN = confusion[0, 0]
FP = confusion[0, 1]
FN = confusion[1, 0]

print(TP)
print(TN)
print(FP)
print(FN)
```

 <b>2b.vii) Deriving Metrics from Confusion Matrix</b>

Using the values derived from the confusion matrix (in part 2b.vi), we can then calculate the following:
- Accuracy 
- Error
- Sensitivity
- Specificity
- Precision

Classification Accuracy: This measure how often is the classifier correct?

```python
# accuracy
# Method 1 (which was covered in part 2b.v)
print(metrics.accuracy_score(y_test, y_pred))

# Method 2: Use the values tabulated from confusion matrix to calculate the accuracy
print((TP + TN) / float(TP + TN + FP + FN))
```

Classification Error: Overall, how often is the classifier incorrect?

```python
# classification error
# Method 1: Using the accuracy derived from accuracy_score(), we minus it against 1
print(1 - metrics.accuracy_score(y_test, y_pred))

# Method 2: Use the values tabulated from confusion matrix to calculate the error
classification_error = (FP + FN) / float(TP + TN + FP + FN)
```

Sensitivity: When the actual value is positive, how often is the prediction correct?

```python
# sensitivity = recall
# Method 1: Use recall_score() function
print(metrics.recall_score(y_test, y_pred))

# Method 2: Use the values tabulated from confusion matrix to calculate the sensitivity
sensitivity = TP / float(FN + TP)
print(sensitivity)
```

Specificity: When the actual value is negative, how often is the prediction correct?

```python
specificity = TN / (TN + FP)
print(specificity)
```

Precision: When a positive value is predicted, how often is the prediction correct?

```python
# Method 1: Use precision_score() function
print(metrics.precision_score(y_test, y_pred))

# Method 2: Use the values tabulated from confusion matrix to calculate the precision
precision = TP / float(TP + FP)
print(precision)
```

<b>2b.vii) Validation by Measuring Accuracy using Train-Test Split</b>

To perform train-test split on logistic regression:
- we can use the <code>train_test_split()</code> to split the data
- pass in the values into <code>metrics.accuracy_score()</code>

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

# Split data into predictors X and output Y
features = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin','BMI', 'DiabetesPedigreeFunction', 'Age']
X = insurance_diabetes_df[features] # Print x out if you don't understand what data it contains
y = insurance_diabetes_df['Outcome']

# Split the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)


# print(X_train) # see for yourself
# print(X_test) # see for yourself

# Instantiate model object and fit model
log_reg = LogisticRegression(max_iter=300)
log_reg_model = log_reg.fit(X_train, y_train)

# Perform prediction with the model
y_pred = log_reg_model.predict(X_test)

# Then we pass in the test outcome and the predicted outcome to get the accuracy score
print(metrics.accuracy_score(y_test, y_pred))
```

<b>2b.viii) Validation by Measuring Accuracy using Cross Validation (k-fold cross validation)</b>

To perform cross validation on logistic regression:

- we can use the <code>KFold()</code> to split the data
- iterate through each of the kfold
- calculate the <b>Accuracy</b> for each of the fold, and append to a list

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import KFold
import statistics

# Split data into predictors X and output Y
features = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin','BMI', 'DiabetesPedigreeFunction', 'Age']
X = insurance_diabetes_df[features] # Print x out if you don't understand what data it contains
y = insurance_diabetes_df['Outcome']

# Split the data into ten set
kf = KFold(n_splits=10)

# Create logistic regression object
log_reg = LogisticRegression(max_iter=300)

# List of accuracy for each fold
k_fold_accuracy = []

# Iterate through each fold and calculate the RMSE for each fold
for train_index, test_index in kf.split(X):
    
    # Extract the training and test data
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]
    
    # Fit model
    log_reg_model = log_reg.fit(X_train, y_train)
    y_pred = log_reg_model.predict(X_test)
    
    #Calculate Accuracy for the fold and append it
    accuracy = metrics.accuracy_score(y_test, y_pred)
    k_fold_accuracy.append(accuracy)
    
print('The accuracy for each fold is:', k_fold_accuracy)
print('The average accuracy is:', statistics.mean(k_fold_accuracy))
```

<b>2b.ix) ROC Curve & AUC Curve (Obtaining the best threshold)</b>

Earlier, we mentioned that in <b>part 2.iv that we should not just arbitrarily assign a threshold value to the model</b>. Indeed, a more systematic approach would be to make use of an ROC curve.

To refresh your memory, earlier we trained the model with the code below:

```python
from sklearn.preprocessing import binarize

# instantiate model
log_reg = LogisticRegression(max_iter=300)

# Earlier when trying to visualise the scatterplot we changed the outcome to str. Here we change it back!
insurance_diabetes_df['Outcome'] = insurance_diabetes_df['Outcome'].apply(int)

# Split data into predictors X and output Y
features = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin','BMI', 'DiabetesPedigreeFunction', 'Age']
X = insurance_diabetes_df[features] # Print x out if you don't understand what data it contains
y = insurance_diabetes_df['Outcome']

# fit model
log_reg.fit(X, y)

# Generate predictions and format results
y_pred_probabilities = log_reg.predict_proba(X)
y_pred_probabilities_formatted = y_pred_probabilities[:, 1]
```

Using the results above, we can then generate a ROC curve:

- First, we use the <code>roc_curve</code> function to obtain the: (1) false positive rate, (2) true positive rate, and (3) thresholds
- Secondly, we pass the parameters <code>px.area()</code> to plot the chart

You can read up more about visualizing ROC curve in this link: https://plotly.com/python/roc-and-pr-curves/

```python
import plotly.express as px
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_curve, auc
from sklearn.datasets import make_classification

# Calculate the (1) false positive rate, (2) true positive rate, and (3) thresholds
y_pred_probabilities_formatted = list(y_pred_probabilities_formatted) # converting to list
fpr, tpr, thresholds = roc_curve(y, y_pred_probabilities_formatted)

# Plotting the chart
fig = px.area(
    x=fpr, y=tpr,
    title=f'ROC Curve (AUC={auc(fpr, tpr):.4f})',
    labels=dict(x='False Positive Rate', y='True Positive Rate'),
    width=700, height=500
)

# This part is just for formatting & adding the dash-line 
fig.add_shape(
    type='line', line=dict(dash='dash'),
    x0=0, x1=1, y0=0, y1=1
)
fig.update_yaxes(scaleanchor="x", scaleratio=1)
fig.update_xaxes(constrain='domain')


fig.show()
```

<div class="alert alert-block alert-warning">
For plotly, we can easily create the ROC curve and AUC curve together in one plot. Visually, we can observe that the the optimal threshold is approxmiately <b>0.3338695381427648</b>. And with this, you can build another model with the optimal threshold.
</div>

<b>3b.vii) Validation by Measuring Accuracy using Cross Validation (k-fold cross validation)</b>

To perform cross validation on KNN:

- we can use the <code>KFold()</code> to split the data
- iterate through each of the kfold
- calculate the <b>Accuracy</b> for each of the fold, and append to a list

```python
from sklearn.tree import DecisionTreeRegressor  
from sklearn.model_selection import KFold
import statistics

# Split data into predictors X and output Y
features = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin','BMI', 'DiabetesPedigreeFunction', 'Age']
X = insurance_diabetes_scaled_df[features] # Print x out if you don't understand what data it contains
y = insurance_diabetes_scaled_df['Outcome']

# Split the data into ten set
kf = KFold(n_splits=10)

# Create KNN model object
knn = KNeighborsClassifier(n_neighbors=3)

# List of RMSE for each fold
k_fold_Accuracy = []

# Iterate through each fold and calculate the RMSE for each fold
for train_index, test_index in kf.split(X):
    
    # Extract the training and test data
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]
    
    # Fit model
    knn_model = knn.fit(X_train, y_train)
    y_pred = knn_model.predict(X_test)
    
    #Calculate RMSE for the fold and append it
    accuracy = metrics.accuracy_score(y_test, y_pred)
    k_fold_Accuracy.append(accuracy)
    
print('The Accuracy for each fold is:', k_fold_Accuracy)
print('The average accuracy is:', statistics.mean(k_fold_Accuracy))
```

<b>3b.viii) Finding the Optimal K Value</b>

Earlier, we mentioned  in <b>part 3b.ii that we should not just arbitrarily assign a threshold value to the model</b>. We need a more systematic approach. The idea is to iteratively train KNN models with different k values, and then test the accuracy of each model to determine the optimal K-Value. Here we can use the following methods to do so:

- Train-test split (we shall showcase this method in the following example)
- K-Fold cross validation

```python
from sklearn.neighbors import KNeighborsClassifier

# Split data into predictors X and output Y
features = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin','BMI', 'DiabetesPedigreeFunction', 'Age']
X = insurance_diabetes_scaled_df[features] # Print x out if you don't understand what data it contains
y = insurance_diabetes_scaled_df['Outcome']

# Split the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)

test_scores = []

for i in range(1,30):
    
    ## Create model object based on iteratively increasing k value
    knn = KNeighborsClassifier(n_neighbors=i)
    
    ## Fit the data on the model with a diff k value
    knn_model = knn.fit(X_train,y_train)
    
    ## Generate prediction on the test dataset
    y_pred = knn_model.predict(X_test)
    
    ## Calculate accuracy score
    accuracy = metrics.accuracy_score(y_test, y_pred)
    
    ## Store accuracy
    test_scores.append(accuracy)
    
## We print the test scores
for i in range(len(test_scores)):
    
    print('K-Value',i+1, test_scores[i])
```

<div class="alert alert-block alert-warning">
Through the above method, we can observe that a k-value of 13 is one of the most optimal k-value. From this, we hope you take away the fact that the model evaluation procedures are not limited to just verifying model accuracy. You can also use them to tune various aspects of your model.</div>

<b>Overfitting in Machine Learning</b>

- Overfitting happens when a model learns the detail and noise in the training data to the extent that it negatively impacts the performance of the model on new data. This means that the noise or random fluctuations in the training data is picked up and learned as concepts by the model. The problem is that these concepts do not apply to new data and negatively impact the models ability to generalize.

- Overfitting is more likely with nonparametric and nonlinear models that have more flexibility when learning a target function. As such, many nonparametric machine learning algorithms also include parameters or techniques to limit and constrain how much detail the model learns.

- For example, decision trees are a nonparametric machine learning algorithm that is very flexible and is subject to overfitting training data. This problem can be addressed by pruning a tree after it has learned in order to remove some of the detail it has picked up.

<b>Underfitting in Machine Learning</b>

- Underfitting refers to a model that can neither model the training data nor generalize to new data.

- An underfit machine learning model is not a suitable model and will be obvious as it will have poor performance on the training data.

- Underfitting is often not discussed as it is easy to detect given a good performance metric. The remedy is to move on and try alternate machine learning algorithms. Nevertheless, it does provide a good contrast to the problem of overfitting.

<b>How to Limit Overfitting</b>

- Both overfitting and underfitting can lead to poor model performance. But by far the most common problem in applied machine learning is overfitting.

- Overfitting is such a problem because the evaluation of machine learning algorithms on training data is different from the evaluation we actually care the most about, namely how well the algorithm performs on unseen data.

- There are two important techniques that you can use when evaluating machine learning algorithms to limit overfitting:
  - Use a resampling technique to estimate model accuracy.
  - Hold back a validation dataset.

- The most popular resampling technique is k-fold cross validation. It allows you to train and test your model k-times on different subsets of training data and build up an estimate of the performance of a machine learning model on unseen data.

- A validation dataset is simply a subset of your training data that you hold back from your machine learning algorithms until the very end of your project. After you have selected and tuned your machine learning algorithms on your training dataset you can evaluate the learned models on the validation dataset to get a final objective idea of how the models might perform on unseen data.

- Using cross validation is a gold standard in applied machine learning for estimating model accuracy on unseen data. If you have the data, using a validation dataset is also an excellent practice.
