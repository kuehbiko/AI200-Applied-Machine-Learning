<b>Write your code below</b>

```python
## Write your code below
# dataset needs: at least 2 categories and 2 aggregate columns
# group categories according to 1 aggregate column, use 2nd aggregate column for continuous color bar
# dataset: https://www.kaggle.com/datasets/utkarshx27/2021-startups
df = pd.read_csv('data/Startups in 2021 end.csv')
df.head()
```

```python
df['Valuation ($B)'] = df['Valuation ($B)'].replace(r'[^\w\s]', '', regex=True)
df['Valuation ($B)'] = df['Valuation ($B)'].astype('int')
```

```python
df['Industry'] = df['Industry'].str.lower().str.replace('finttech', 'fintech')
```

```python
df['Industry'].value_counts()
```

```python
df['startup_count'] = df['Industry']
```

```python
treemap_df = df.groupby(['Industry']).agg({'startup_count':'count', 'Valuation ($B)':'mean'}).reset_index()
treemap_df.head(10)
```

```python
import numpy as np

# specclass - count - pop
# key level - average - lifeexp
fig = px.treemap(treemap_df, path=[px.Constant('All'), 'Industry'], 
                 values='startup_count', color='Valuation ($B)', 
                 color_continuous_scale='viridis_r',
                 color_continuous_midpoint=np.average(treemap_df['Valuation ($B)'], weights=treemap_df['startup_count'])
                )

fig.show()
```

<b> Write your interpretation of the visualisation below</b>


```python
'''
by quantity: 
fintech and internet & software services have the largest count of startups over time
consumer & retail and travel have the least

by valuation: 
auto & transportation have the highest median valuation (in billions) out of all startups,
mobile & telecommunications have the lowest

auto & transportation startups might be a lucrative job market due to its small size but high value
'''
```
