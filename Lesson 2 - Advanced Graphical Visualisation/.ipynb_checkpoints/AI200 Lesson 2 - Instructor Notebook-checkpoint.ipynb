3d. Customising your Visualisations with Plotly.express
---

To add richness to our visualisation, we can customise our visualisation using plotly.express's in-built parameters and functions. In this section, we cover the commonly used customisations that applies to most plotly visualisations:

i) Axes <br>
ii) Data colour & label <br>
iii) Trendlines <br>
iv) Title <br>
v) Facet / Trellis Plots  (Aka subplots in Matplotlib) <br>
vi) Marginal Plots<br>
vii) Background Color

In general, you can find out more about styling your visualisations in this page: https://plotly.com/python/styling-plotly-express/

<b>3d.i) Axes</b>

Axis titles are automatically set to the column names when using Plotly Express with a data frame as input. However, if you want to customise the axes, you can use the <code>update_xaxes()</code> and <code>update_yaxes()</code> functions. In fact, you can use these functions to add a prefix to the value in the axes by passing in the <code>tickprefix</code> parameter.

<br>You can read more at:  https://plotly.com/python/axes/

```python
fig = px.scatter(creditcard_spending_df,
                 x='total_credit_card_bill',
                 y='amount_spent_on_alcohol'
)

# This is how we modify the axes title and add a prefix to the axes value
fig.update_yaxes(title_text="Credit Card Spending on Alcohol", tickprefix="$")
fig.update_xaxes(title_text="Total Credit Card Bill", tickprefix="$")

fig.show()
```

_Try moving your mouse over some of the data points to see the "$" prefix update as well!_ 

<b>3d.ii) Data Color and Labels</b>

You can use the <code>color</code> parameter, and pass the name of the column whose values you want to use to assign a colour to the data. To have discrete colors, note that the data in the column should be **categorical** instead of numerical.

<br>You can read more at:  https://plotly.com/python/axes/

```python
# Add 'color' parameter to assign colours to datapoints based on customer gender
fig = px.scatter(creditcard_spending_df,
                 x='total_credit_card_bill',
                 y='amount_spent_on_alcohol',
                 color='sex'
)

fig.update_yaxes(title_text="Amount Spent on Alcohol", tickprefix="$")
fig.update_xaxes(title_text="Total Bill", tickprefix="$")

fig.show()
```

<div class="alert alert-block alert-warning">
Isn't plotly powerful? With just one parameter change, you can instantly study the difference in behaviour between male and female clients.
</div>

To change the name of the label and legend, you can use the <code>label</code> parameter and pass in a dictionary (with the key being the current default label, and value being the name you want to change it to)

```python
# Add 'labels' parameter to customise the text of the labels
fig = px.scatter(creditcard_spending_df, 
                 x = 'total_credit_card_bill', y = 'amount_spent_on_alcohol',
                 color = 'sex',
                 labels = {"sex": "Gender", "total_credit_card_bill": "Total Credit Card Bill", "amount_spent_on_alcohol": "Credit Card Spending on Alcohol"}
)

fig.update_yaxes(tickprefix="$")  # Note that title_text can be excluded once you have labels
fig.update_xaxes(tickprefix="$")

fig.show()
```

<b>3d.iii) Trendlines</b>

You can add trendlines to show the trend of the data using the <code>trendline</code> parameter. Later in your machine learning class on regression, you will learn the significance and mechanism of ordinary least squares (ols).

<br>You can read more at: https://plotly.com/python/linear-fits/

```python
# Add 'trendline' parameter to fit a trendline using Ordinary Least Squares method
fig = px.scatter(creditcard_spending_df, 
                 x = 'total_credit_card_bill', y = 'amount_spent_on_alcohol',
                 color = 'sex',
                 labels = {"sex": "Gender", "total_credit_card_bill": "Total Credit Card Bill", "amount_spent_on_alcohol": "Credit Card Spending on Alcohol"},
                 trendline = 'ols'  # add trendline parameter
)

fig.update_yaxes(tickprefix="$")
fig.update_xaxes(tickprefix="$")

fig.show()
```

<b>3d.iv) Title</b>

To add a title to your graph, you can simply use the <code>title</code> parameter.

<br>You can read more at: https://plotly.com/python/figure-labels/

```python
# Add 'title' parameter to insert a title for the figure
fig = px.scatter(creditcard_spending_df, 
             x = 'total_credit_card_bill', y = 'amount_spent_on_alcohol',
             color = 'sex',
             labels = {"sex": "Gender", "total_credit_card_bill": "Total Credit Card Bill", "amount_spent_on_alcohol": "Credit Card Spending on Alcohol"},
             trendline = 'ols',
             title = 'Amount Spent on Alcohol (Male vs Females)' # add title parameter
)

fig.update_yaxes(tickprefix="$")
fig.update_xaxes(tickprefix="$")

fig.show()
```

<b>3d.v) Facet / Trellis Plots</b>

Facet plots, also known as trellis plots or small multiples, are figures made up of multiple subplots which have the same set of axes, where each subplot shows a subset of the data. If you have taken AI100, you will realise that this is similar to matplotlib's subplot. In Plotly, we can create multiple plots in the same figure through two parameters:
- <code>facet_row</code> or,
- <code>facet_col</code>

When you use <code>facet_row</code>:
- the data will be sliced into different subsets based on the feature indicated for <code>facet_row</code>. Note that each unique value of the feature will be sliced into a subset on its own (e.g. smoker=No, smoker=Yes)
- thereafter, the different subsets of data will be used in their respective graphs, and these graphs will be <b>stacked row by row</b> in the same figure

<br>You can read more at: https://plotly.com/python/facet-plots/

```python
# Add 'facet_row' parameter to create facet plot (stacked vertically)
fig = px.scatter(creditcard_spending_df, 
             x = 'total_credit_card_bill', y = 'amount_spent_on_alcohol',
             color = 'sex',
             labels = {"sex": "Gender", "total_credit_card_bill": "Total Bill", "amount_spent_on_alcohol": "Spending on Alcohol"},
             trendline = 'ols',
             title = 'Amount Spent on Alcohol (Male vs Females)',
             facet_row = 'smoker' # add facet_row parameter
)

fig.update_yaxes(tickprefix="$")
fig.update_xaxes(tickprefix="$")

fig.show()
```

<div class="alert alert-block alert-warning">
Those who have used matplotlib before, or even other visualisation libraries will realise how much easier it is to generate multiple plots in the same figure with plotly!
</div>

When you use <code>facet_col</code>:
- the data will be sliced into different subsets based on the feature indicated for <code>facet_col</code>. Note that each unique value of the feature will be sliced into a subset on its own (e.g. smoker=No, smoker=Yes)
- thereafter, the different subsets of data will be used to their respective graphs, and these graphs will be <b>stacked side by side</b> in the same figure

```python
# Add 'facet_col' parameter to create facet plot (stacked horizontally)
fig = px.scatter(creditcard_spending_df, 
             x = 'total_credit_card_bill', y = 'amount_spent_on_alcohol',
             color = 'sex',
             labels = {"sex": "Gender", "total_credit_card_bill": "Total Credit Card Bill", "amount_spent_on_alcohol": "Credit Card Spending on Alcohol"},
             trendline = 'ols',
             title = 'Amount Spent on Alcohol (Male vs Females)',
             facet_col = 'smoker' # add facet_row parameter
)

fig.update_yaxes(tickprefix="$")
fig.update_xaxes(tickprefix="$")

fig.show()
```

In the event that your feature has a large number of unique values, it is possible to wrap columns using the facet_col_wrap argument. You can augment your code with the <code>facet_col_wrap</code> parameter to constrain the graphs. The number which you pass to the <code>facet_col_wrap</code> will dictate the maximum number of columns of graph in a row.

```python
# Add 'facet_col_wrap' parameter to wrap the facet plots
fig = px.scatter(creditcard_spending_df, 
             x = 'total_credit_card_bill', y = 'amount_spent_on_alcohol',
             color = 'sex',
             labels = {"sex": "Gender", "total_credit_card_bill": "Total Credit Card Bill", "amount_spent_on_alcohol": "Spending on Alcohol"},
             trendline = 'ols',
             title = 'Amount Spent on Alcohol (Male vs Females)',
             facet_col = 'usual_drinking_day',
             facet_col_wrap = 2, # add facet_col_wrap parameter
)

fig.update_yaxes(tickprefix="$")
fig.update_xaxes(tickprefix="$")

fig.show()
```

_For further exploration:_ More facet plot examples [here](https://plotly.com/python/facet-plots/#histogram-facet-grids).

<b>3d.vi) Marginal Plots</b>

Certain graphs like scatter plot have unique features that allows you to add  add other kinds of plot to enrich the main graphical visualisation. In the case of scatter plot we can do so using these parameters:
- <code>marginal_x</code> - One of 'rug', 'box', 'violin', or 'histogram'. If set, a horizontal subplot is <b>drawn above the main plot, visualizing the x-distribution</b>.
- <code>marginal_y</code> - One of 'rug', 'box', 'violin', or 'histogram'. If set, a vertical subplot is <b>drawn to the right of the main plot, visualizing the y-distribution</b>.

<br>You can read more at: https://plotly.com/python-api-reference/generated/plotly.express.scatter.html#

```python
# Add marginal_x and marginal_y parameter to add auxiliary plots
fig = px.scatter(creditcard_spending_df, 
             x = 'total_credit_card_bill', y = 'amount_spent_on_alcohol',
             color = 'sex',
             labels = {"sex": "Gender", "total_credit_card_bill": "Total Credit Card Bill", "amount_spent_on_alcohol": "Credit Card Spending on Alcohol"},
             trendline = 'ols',
             title = 'Amount Spent on Alcohol (Male vs Females)',
             marginal_x = 'box',
             marginal_y = 'violin', # add marginal_x and marginal_y parameter
)

fig.update_yaxes(tickprefix="$")
fig.update_xaxes(tickprefix="$")

fig.show()
```

<div class="alert alert-block alert-warning">
Now you can instantly compare the distrbution of the amount spent on alcohol as well as the bill amount for both male and female clients!
</div>

<b>3d.vii) Background Color</b>

Finally, if you want to change the background color of the plots, an easy way to do it is via the <code>template</code> parameter. You can check out the list of possible colors which you can pass in to the <code>template</code> parameter at: https://plotly.com/python/templates/


```python
# Add template parameter to change the color scheme of the plot
fig = px.scatter(creditcard_spending_df, 
             x = 'total_credit_card_bill', y = 'amount_spent_on_alcohol',
             color = 'sex',
             labels = {"sex": "Gender", "total_credit_card_bill": "Total Credit Card Bill", "amount_spent_on_alcohol": "Credit Card Spending on Alcohol"},
             trendline = 'ols',
             title = 'Amount Spent on Alcohol (Male vs Females)',
             marginal_x = 'box',
             marginal_y = 'violin',
             template = 'simple_white' # added the template parameter
)

fig.update_yaxes(tickprefix="$")
fig.update_xaxes(tickprefix="$")

fig.show()
```

4c. Visualisations for Relationship
---

For this category of visualisation, it involves trying to <b>establish a relationship between different features of a dataset</b>. Loosely put, we are trying to <b>compare the relationship between multiple columns in a dataset</b>.

Types of charts under this category include:<br>

<b>4c.i) Scatterplot</b>

In the earlier <b>credit card spending dataset</b>, we used a scatterplot to establish the relationship between the credit card bill, and amount spend on alchohol. It was evident from the chart there was a positive relationship between the credit card bill and amount spent on alcohol (meaning: As the credit card bill increases, there is a strong likelihood of an increase in amount spent on alchohol). In sum, a <b>scatterplot is great when you are trying to understand the relationship between 2 features / 2 columns in a dataset at any one point in time</b>. 

<b>Common use cases:</b>
- establishing relationship between Height and weight
- establishing relationship between Floor size and house price
- establishing relationship between Features (for smaller datasets, you can use scatterplot to help select the right features to build your machine learning model)

You can read more at: https://plotly.com/python/line-and-scatter/

```python
# Printing the data out first so you can see the name of the available features
creditcard_spending_df.head()
```

```python
# Same code from previous section on customising parameters
fig = px.scatter(creditcard_spending_df, 
             x = 'total_credit_card_bill', y = 'amount_spent_on_alcohol',
             color = 'sex',
             labels = {"sex": "Gender", "total_credit_card_bill": "Total Credit Card Bill", "amount_spent_on_alcohol": "Credit Card Spending on Alcohol"},
             trendline = 'ols',
             title = 'Amount Spent on Alcohol (Male vs Females)' # add title parameter
)

fig.show()
```

<b>4c.ii) Scatterplot Matrices</b>

If you want to<b> understand the relationship between all the pairs of features in a dataset in a single snapshot, you can make use of a scatterplot matrices</b>. To create a scatterplot matrices, you make use of the <code>scatter_matrix()</code> function, and you pass in a list of the names features/columns which you want to study the relationship between to the <code>dimension</code> parameter. In fact, you can also use the <code>color</code> parameter in this chart as well. Let's see an example using the same <b>credit card spending dataset</b>!<br>
     
<b>Common use cases:</b>
- This plot is often used by Data Scientists / Data Analysts to <b>study the relationship between features & the outcome (we are trying to predict)</b>. This helps in the selection of features for building a machine learning model. We will learn more about this from session 3 onwards.
     
You can read more at: https://plotly.com/python/splom/

```python
# Display data first to review useful features
creditcard_spending_df.head()
```

Notice that we pass in a list of features to the <code>dimensions</code> parameter, which are then plotted in a scatterplot against each other. Now, <b>in one snapshot, you can now study the relationship of multiple pairs of features.</b>

```python
creditcard_spending_df.columns
```

```python
fig = px.scatter_matrix(creditcard_spending_df, dimensions=["total_credit_card_bill", "amount_spent_on_alcohol", "drinking_group_size", "usual_drinking_day"],
                        labels = {"sex": "Gender", "total_credit_card_bill": "Total Bill", "amount_spent_on_alcohol": "Alcohol", "drinking_group_size": "Grp Size", "usual_drinking_day": "Day"},
                        color="smoker")
fig.show()
```

<b>4c.iii) Heatmap</b>

The heatmap is another tool that helps us<b> understand the relationship between all the pairs of features in a dataset</b>.
     
<b>Common use cases:</b>
- This plot is often used by Data Scientists / Data Analysts to <b>study the relationship between features & the outcome (we are trying to predict)</b>. This helps in the selection of features for building a machine learning model. We will learn more about this from session 3 onwards.
     
You can read more at: https://plotly.com/python/heatmaps/

Remember in our previous class in Pandas we learnt how to use the <code>corr()</code> function to calculate the relationship of all variables in a dataset? You can learn more about this function in the documentation <a href="https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.corr.html">here</a>.<br>

Let's apply it to the <code>creditcard_spending_df</code>.

```python
corr = creditcard_spending_df.corr(numeric_only=True)
```

Now, let's represent the correlation using a heatmap. All we need to do is just use the <code>imshow()</code> function, and pass in a dataframe of the values which we want to cast onto the heatmap. Here we added two parameters:

- <code>color_continuous_scale</code> - which allows you to customise the color scheme of the heatmap
- <code>text_auto</code> - which allows you to enable the display of the correlation value on each of the boxes in the heatmap (try changing it to false and see what happens)

```python
fig = px.imshow(corr, color_continuous_scale = 'Brwnyl', text_auto = True)
fig.update_xaxes(side="top")
fig.show()
```

<div class="alert alert-block alert-warning">
Note that the heatmap's application is not just restricted to helping us studying the correlation of variables for machine learning. In fact, here is another well established use case for heatmap. <a href='https://medium.com/@marketahead.com/day-month-performance-heatmap-stock-data-analysis-8d5d981e0930'>Stock Analysis!</a>

<b>4c.iv) Bubble Chart</b>

While a <b>scatterplot is great when you are trying to understand the relationship between 2 features / 2 columns in a dataset at any one point in time</b>, it does not work when you are trying to measure the relationship of 3 variables at the same time. Here is where a bubble chart shines! Let's see an example using the data containing the <b>development statistics of countries</b>! In this example, we are trying to understand the relationship between <b>GDP</b>, <b>Life expectancy</b>, and <b>population size</b> for year <b>2007</b>!

<b>Common use cases:</b>
- Visualise high-dimensional data e.g. comparing multiple metrics across countries
     
You can read more at: https://plotly.com/python/splom/

```python
# Display data first to review useful features
countries_df = px.data.gapminder()
countries_df.head(20)
```

Before we begin, we need to make use of <b>conditional selectors in Pandas</b> to extract all the data for 2007.

```python
# Remember your pandas data tranformation / manipulation from the previous session?
countries_2007_df = countries_df[countries_df.year==2007]
countries_2007_df
```

Implementation wise, in Plotly, <b>Bubble chart</b> is also generated through the <code>scatter()</code> function. However, we need to pass in the <code>size</code> parameter. In the example below, by passing in <code>size='pop'</code>, we are instructing plotly to vary the size of each dot in proportion to the value of the feature <b>'pop'</b> (which is basically the population). There are two other parameters in the sample code which we have not introduced before:

- <code>log_x</code> When this parameter is set to <code>True</code>, you are instructing plotly to set the x-axis to a logarithmic scale (as a natural extention, to set the y-axis to a logarithmic scale you can use <code>log_y=True</code>)
- <code>size_max</code> This parameter limits the max size of each dot based on the provided inpu


```python
fig = px.scatter(countries_2007_df, x="gdpPercap", y="lifeExp", size="pop", color="continent",
                 hover_name="country", log_x=True, size_max=60)
fig.show()
```

<b>4c.v) Animated Bubble Chart</b>

We can make use of the <code>animation_frame</code> parameter to animate our bubble chart. This is particularly useful when we are trying to studying <b>the change in relationship of 3 features over time</b>.

<b>Common use cases:</b>
- Visualise high-dimensional data e.g. comparing multiple metrics across countries
- Especially if you are trying to show highlight changes over time, this chart will be immensely useful

You can read more at: https://plotly.com/python/animations/

```python
# Display data first to review useful features
countries_df.head()
```

There are two parameters which you will use to generate the animation:

- <code>animation_frame</code> This set the values for each animation frame. For example, by setting <code>animation_frame="Year"</code>, we are instructing Plotly to use vary each frame by the "year" feature in the dataframe.
- <code>animation_group</code> This sets the values for grouping data – if available.


```python
fig = px.scatter(countries_df, x="gdpPercap", y="lifeExp", size="pop", color="continent", 
                 hover_name="country", log_x=True, size_max=45, 
                 animation_frame="year", animation_group="country", # add animation_frame and animation_group
                 range_x=[100,100000], range_y=[25,90]              # define x & y axis range for animation duration
)
fig.show()
```

<div class="alert alert-block alert-warning">
Notice that with the animated chart, it becomes very clear how immense and rapid the growth in Asia really was! And through this, you should begin to appreciate how useful this chart is when you are trying to <b>bring attention to changes in features over time</b>.
</div>

We can also make use of the <code>facet_col</code> parameter we learnt earlier to further seggregate each plot based on their continent!

```python
fig = px.scatter(countries_df, x="gdpPercap", y="lifeExp", size="pop", color="continent", 
                 hover_name="country", log_x=True, size_max=45, 
                 animation_frame="year", animation_group="country",
                 range_x=[100,100000], range_y=[25,90],
                 facet_col="continent" # add new parameter: facet_col
)
fig.show()
```

<b>4d.iv) Stacked Column Chart</b>

Stacked Column Charts are great for part-to-whole comparisons across categories and even across time (stacked area chart is better for time though). Let's see an example using some psudeo sales data we created below for you to appreciate how to apply this visualisation technique in a business setting!

<b>Common use cases:</b>
- Sales of different categories of products on different sales channel
- Demography of population in different countries

You can read more at: https://plotly.com/python/bar-charts/

```python
## In this cell, we create a dataframe of pseudo sales data for demonstration purposes
## Don't concern yourself too much with the code below!

import numpy as np
import pandas as pd

bank_division = pd.Series(['Retail Banking','Commercial Banking','Private Banking']*12)

series_list = [pd.Series([month]*3) for month in ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']]
month_col = pd.concat(series_list, ignore_index=True)

sales_revenue= [[np.random.randint(5,20), np.random.randint(5,8), np.random.randint(30,50)] for i in range(0,12)]
sales_revenue = sum(sales_revenue, [])

bank_revenue_df = pd.DataFrame({
    'month': month_col,
    'year': 2020,
    'bank division': bank_division,
    'sales revenue ($ Mil)': sales_revenue
})
bank_revenue_df
```

Simple use the <code>bar()</code> function, and specify the features from the dataframe which you want to use for your <code>x</code>, <code>y</code> and the <code>color</code>.

```python
fig = px.bar(bank_revenue_df, x="month", y="sales revenue ($ Mil)", color="bank division", 
             title="Bank Revenue in 2020"
)
fig.show()
```

<b>4d.iv) Stacked Area Chart</b>

Stacked Area Charts are great for part-to-whole comparisons, and also for showing how values develop over time. Let's see an example using the <code>countries_df</code> dataset which was used earlier!

For stacked area charts, you just need to use the <code>area()</code> function, and specify the features from the dataframe which you want to use for your <code>x</code>, <code>y</code> and the <code>color</code>.

```python
fig = px.area(countries_df, x="year", y="pop", color="continent",line_group="country")
fig.show()
```

4e. Visualisations for Composition (Static)
---
For this category of visualisation,it involves trying to <b>purely illustrate the composition of your data</b>. Here, the focus is on the percentage/proportion of each category relative to each other with zero regard for what happens over time.

Under this category, we will cover:<br>
i) Pie Chart <br>
ii) Treemap (not featured in graph icon) <br>

<b>4e.i) Pie Chart</b>

A pie chart is best used when trying to work out the composition of something. If you have categorical data then using a pie chart would work really well as each slice can represent a different category. Let's see an example using the data containing the <b>development statistics of countries</b>! In this example, let's try to compare the population amongst all countries on the America continent in year 2007!


<b>Common use cases:</b>
- compare areas of growth within a business such as turnover, profit and exposure.
- compare market shares

You can read more at: https://plotly.com/python/pie-charts/

```python
# Display data first to review useful features
import pandas as pd
 
# initialize list of lists
data = [['DBS', 148.2], ['UOB', 125.4], ['OCBC', 95.9], ['SCB', 60.3], ['HSBC', 53.6], ['Maybank', 38.5], ['CIMB', 29], ['Citibank', 17.1], ['Others', 67.5]]
 
# Create the pandas DataFrame
domestic_loans_df = pd.DataFrame(data, columns = ['Financial Institute', 'Domestic Loan by Market Share in bn (2017)'])
 
# print dataframe.
domestic_loans_df
```

For stacked area charts, you just need to use the <code>pie()</code> function, and pass the name of the column which you want to visualise in a pie chart to the <code>values</code> parameter. You can also make use of the <code>hover_data</code> and <code>labels</code> parameters to add extra elements to the label (the box pops up when you hover your cursor over the pie chart).

```python
fig = px.pie(domestic_loans_df, values='Domestic Loan by Market Share in bn (2017)', names='Financial Institute',
             title='Singapore Domestic Loans by Market Share in 2017',
             hover_data=['Domestic Loan by Market Share in bn (2017)'])

fig.update_traces(textposition='inside', textinfo='percent+label')
fig.show()
```

<b>4e.ii) Treemap</b>

A treemap is used to show the composition of a whole when there are many components. The arrangement is typically (but not always) a rectangular area divided up into smaller rectangles to represent sub-categories. The size of these sub-category rectangles represents a quantitative value. 

Treemaps can work well if your data falls into this scenario:

- You want to visualize a part-to-whole relationship amongst a large number of categories.
- Precise comparisons between categories is not important.
- The data is hierarchical.

Let's see an example using the data containing the <b>development statistics of countries</b>! In this example, let's try to compare the life expectancy amongst all countries in year 2007!

<b>Common use cases:</b>
- Compare effectiveness of sales channels in different countries
- Compare market shares
- Showcase gains and losses of stockmarket [Often used by S&P500] (https://finviz.com/map.ashx)
- US electoral college votes

You can read more at: https://plotly.com/python/treemaps/

```python
# Display data first to review useful features
countries_df.head()
```

Before we begin, we need to use <b>conditional selectors in Pandas</b> to extract all the data for all countries in 2007.

```python
countries_2007_df = countries_df[countries_df.year==2007]
countries_2007_df.head()
```

For treemaps, you just need to use the <code>treemap()</code> function, and pass the order of hierarchy in a list to <code>path</code>. You must indicate the column which you want to visualise to the <code>values</code> parameter. <br>

Optionally, you can also modify the <code>hover_template</code> to customise the box that pops up when you hover your cursor over the treemap.

```python
fig = px.treemap(countries_2007_df, values='pop', color='lifeExp', 
                 color_continuous_scale='RdBu',
                 path=[px.Constant('World'), 'continent', 'country']
)
fig.data[0].hovertemplate = '%{label}<br>%{value}'
fig.show()
```

<div class="alert alert-block alert-warning">
Many people tend to misuse treemap, using it in inappropriate cases just because it is sexy and they want to add some visual variety to their analysis. However, when it is used properly, it can be a very powerful tool that is superior to other visualisation techniques. Read this article for an example of a well-applied heatmap (be sure to scroll all the way to the bottom)  <a href='https://www.datarevelations.com/resources/in-praise-of-treemaps/'>In praise of Treemap!</a>

<b>4f.i) Histogram / Displot</b>

As its name suggests, Displot provides a visual representation of the underlying frequency distribution (shape) of a set of continuous data. This allows the inspection of the data for its underlying distribution (e.g., normal distribution), outliers, skewness, etc. Let's see an example using the <code>creditcard_spending_df</code> dataset which was used earlier!

You can read more at: https://plotly.com/python/distplot/

```python
# Display data first to review useful features
creditcard_spending_df.head()
```

Simply use the <code>histogram()</code> function, and specify the features from the dataframe which you want to use for your <code>x</code>, <code>y</code> and the <code>color</code>. Additionally, you can use the <code>marginal</code> parameter (accepts <code>box</code>, <code>violin</code> or <code>rug</code>) to add an additional distribution tool on top of the histogram.

```python
import plotly.express as px

fig = px.histogram(creditcard_spending_df, 
                   x = 'total_credit_card_bill',
                   y = 'amount_spent_on_alcohol',
                   color = 'sex',
                   marginal = 'box' # or violin, rug
)

fig.show()
```

<b>4g.i) Line Chart</b>

Line charts are used to track changes over short and long periods of time. When smaller changes exist, line charts are better to use than bar graphs. Line charts can also be used to compare changes over the same period of time for more than one group.

Let's see an example using the <code>stock_df</code> dataset which was used earlier!

```python
import dash
from dash import dcc
from dash import html

app = dash.Dash()
app.layout = html.Div([
    dcc.Graph(figure=line_chart_plot)           # Simply assign your line_chart_plot visualisation to 'figure' parameter
])

app.run_server(debug=True, use_reloader=False)  # Turn off reloader if inside Jupyter
```
